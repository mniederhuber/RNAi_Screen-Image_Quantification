```python
import napari
import numpy as np
import pandas as pd
import aicsimageio
```

```python
sample_sheet = pd.read_csv('sheets/results_fp.csv')
sample_sheet['date'] = sample_sheet['date'].fillna(0).astype(int)
sample_sheet['date'] = sample_sheet['date'].astype(int)

sample_id = []

for x, y in sample_sheet.iterrows():
    #print(x)
    sample_id.append('_'.join([y['Symbol'],y['line'],str(y['date'])]))
    
sample_sheet['id'] = sample_id  
```

```python
images = {}

for sample_id in sample_sheet['id']:
    line = sample_sheet.loc[sample_sheet['id'] == sample_id, 'line'].iloc[0]
    date = sample_sheet.loc[sample_sheet['id'] == sample_id, 'date'].iloc[0]
    fp = sample_sheet.loc[sample_sheet['id'] == sample_id, 'fp'].iloc[0]
    symbol = sample_sheet.loc[sample_sheet['id'] == sample_id, 'Symbol'].iloc[0]
    
    if symbol not in images.keys():
        images[symbol] = [{line : fp}]
    else:
        images[symbol].append({line : fp})

        
images        
```
